## Library imports

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## Data loads

```python
DATA_DIR = "../data/raw/"

# Load dataset metadata (if available)
df = pd.read_csv("../data/raw/metadata.csv")
df.head()
```

## Class distribution

```python
# Class distribution
plt.figure(figsize=(6,4))
sns.countplot(x='label', data=df)
plt.title("Class Distribution")
plt.show()
```

```python



import random
import cv2

sample_images = random.sample(os.listdir(f"{DATA_DIR}/melanoma"), 5)

for img in sample_images:
    img_path = f"{DATA_DIR}/melanoma/{img}"
    img_data = cv2.imread(img_path)
    plt.imshow(img_data)
    plt.title(img)
    plt.axis('off')
    plt.show()
```
